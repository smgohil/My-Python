```python
import random
global Username
class TaskMgmt:
    def __init__(self,filename='task.txt'):
        self.filename=filename

    def addtask(self, taskdetails, status):
        file = open(self.filename,'a') 
        taskid=random.randint(1,99999)
        file.write(f"{taskid}:{taskdetails}:{Username}:{status}\n")
        print('Task added successfully with ID{taskid}')

    def viewtask(self):
        with open(self.filename,'r') as file:
            for line in file:
                stored_taskid, stored_taskdetail,stored_username,stored_taskstatus = line.strip().split(':')
                print('Task Id: '+ stored_taskid)
                print('Task details: ' + stored_taskdetail)
                print('Task Status: ' + stored_taskstatus)
                print('----------------------------------------------------------')

    
    def replaceline(self,oldline,newline):
        with open(self.filename, 'r') as file:
            lines = file.readlines()
            for i in range(len(lines)):
                if lines[i].strip() == oldline:
                    lines[i] = newline + '\n'  
    
        with open(self.filename, 'w') as file:
            file.writelines(lines)

    def deletetask(self,taskid):
        with open(self.filename, 'r') as file:
            for line in file:
                stored_taskid, stored_taskdetail,stored_taskstatus,stored_username = line.strip().split(':')
                oldline = stored_taskid + ":" + stored_taskdetail + ":" + stored_username + ":" + stored_taskstatus
                newline = stored_taskid + ":" + stored_taskdetail + ":" + stored_username + ":" + "Deleted"
                if stored_taskid == taskid: 
                    print("Taskid Found")
                    self.replaceline(oldline,newline)
                    return True
        print("Taskid not found")
        return False

    def completetask(self,taskid):
        with open(self.filename, 'r') as file:
            for line in file:
                stored_taskid, stored_taskdetail,stored_taskstatus,stored_username = line.strip().split(':')
                oldline = stored_taskid + ":" + stored_taskdetail + ":" + stored_username + ":" + stored_taskstatus
                newline = stored_taskid + ":" + stored_taskdetail + ":" + stored_username + ":" + "Completed"
                if stored_taskid == taskid: 
                    print("Taskid Found")
                    self.replaceline(oldline,newline)
                    return True
        print("Taskid not found")
        return False    
class UserAuth:
    def __init__(self, filename='users.txt'):
        self.filename = filename

    def register(self, username, password):
        with open(self.filename, 'a') as file:
            
            if self.is_username_taken(username):
                print("Username already exists. Please choose a different username.")
            else:
                file.write(f"{username}:{password}\n")  
                print("Registration successful!")

    def authenticate(self, username, password):
        with open(self.filename, 'r') as file:
            for line in file:
                stored_username, stored_password = line.strip().split(':')
                if stored_username == username and stored_password == password:
                    print("Authentication successful!")
                    return True
        print("Authentication failed. Invalid username or password.")
        return False

    def is_username_taken(self, username):
        try:
            with open(self.filename, 'r') as file:
                for line in file:
                    stored_username, _ = line.strip().split(':')
                    if stored_username == username:
                        return True
        except FileNotFoundError:
            return False  
        return False
    


#  display the menu
def menu():
    while True:
        print("\nWelcome to Task management system")
        print("Press:")
        print("1 to Add Task")
        print("2 to View Task")
        print("3 to Mark a task to completed")
        print("4 to Delete a Task")
        print("5 to Register a User")
        print("6 to Logout")
        
        ch = input("Please enter your menu choice: ")

        if ch == '1':
           taskdesc = input("Please Enter Task Details")
           
            # Add Task
           Task = TaskMgmt()
           Task.addtask(taskdesc,'pending')
        elif ch == '2':
            print('Task Details Are\n')
            Task=TaskMgmt()
            Task.viewtask()
        elif ch == '3':
            taskid = input("Please enter Taskid to Complete")
            Task=TaskMgmt()
            Task.completetask(taskid)
        elif ch == '4':
            taskid = input("Please enter Taskid to delete")
            Task=TaskMgmt()
            Task.deletetask(taskid)
        elif ch == '5':
           Username = input("Please Enter User Name to register user")
           Password = input("Please Enter Password to register user")
            
           auth_system = UserAuth()
           auth_system.register(Username, Password)
           
        elif ch == '6':
            print("Goodbye!")
            break
        else:
            print("Please Enter Valid option for Menu")

if __name__ == "__main__":
    
    Username = input("Please Enter User Name to Login")
    Password = input("Please Enter Password to Login")
    auth_system = UserAuth()
    if auth_system.authenticate(Username, Password):
        menu()
    else:
        print("User not allowed to login")
        #break


 
```
